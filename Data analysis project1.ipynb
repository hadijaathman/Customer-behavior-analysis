# Checking for missing values

```python
df.isnull().sum()
```

As we can see from the cell above, we have 37 missing values under review rating. This has to be fixed before continuing. we shall go ahead and replace the missing values with the median because meadian is not affected by outlieses as compared to mean. and we shall replace the median of a specific cateory so that the values are assigned according to a given category

# Replacing missing values with median per category

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

checking if the missing values have been replaced

```python
df.isnull().sum()
```

There are no more null values however, looking at the column names, we have mixed lowercase and uppercase, there are also spaces between words and this could be a problem in python and sql. in the next step we shall change this so thall all the column names follow the snake pattern which has all the column names written in lowercase and all spacings replaced with an underscore.

# column name editing to snake casing for easy usage.

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

# Data engineering

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

# creating a new columns to group age group

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

# Column frequency_of_purchase is written in text. this would be more usefull if it were numbered and the text can be converted t  to numbers

```python
frequency_mapping ={
    'Fortnightly': 14,
    'Weekly' : 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
